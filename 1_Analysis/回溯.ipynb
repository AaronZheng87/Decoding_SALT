# 根据自己的需要改n和op_value的值

```python
op_value = "word_first" # 设置要操作的condition值
group = df.groupby("subj_idx", "condition") # 分组
for i in group:
    if i[0] == op_value:
        op_group = i[1]
        n = 1 # 设置要回溯几层
        new_col_name = f"rt_{n}"
        ##print("要回溯的列：")
        ##print(op_group["rt"])
        
        ##print("回溯后：")
        print(op_group["rt"].shift(n))
        op_group[new_col_name] = op_group["rt"].shift(n)
        
        op_group.dropna(inplace=True) # 删除空值对应的行
#         保存修改后的dataframe
        #op_group.to_csv(f"processed_{op_value}.csv", index=False)
```
